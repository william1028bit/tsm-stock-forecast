```python
import yfinance as yf
import ta
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from xgboost import XGBRegressor
from sklearn.model_selection import TimeSeriesSplit, ParameterGrid
from sklearn.metrics import mean_absolute_error, mean_squared_error

%matplotlib inline

# =============================
# 0) 
# =============================
def rmse(a, b):
    return float(np.sqrt(mean_squared_error(a, b)))

def add_features(df: pd.DataFrame) -> pd.DataFrame:
    """ +  fragmented"""
    # TA 
    df = ta.add_all_ta_features(
        df, open="Open", high="High", low="Low", close="Close", volume="Volume", fillna=True
    )

    #  + 
    base = pd.DataFrame(index=df.index)
    base["month"] = df["Date"].dt.month
    base["dayofweek"] = df["Date"].dt.dayofweek
    base["return_1"] = df["Close"].pct_change(1)
    base["return_5"] = df["Close"].pct_change(5)
    base["return_10"] = df["Close"].pct_change(10)
    base["cci"] = ta.trend.cci(df["High"], df["Low"], df["Close"], window=14)
    base["stc"] = df["Close"].rolling(window=5).std()

    #   lag next-day 
    for lag in [1, 2, 5]:
        base[f"open_lag_{lag}"] = df["Open"].shift(lag)
        base[f"close_lag_{lag}"] = df["Close"].shift(lag)
        base[f"vol_lag_{lag}"] = df["Volume"].shift(lag)

    #   rolling
    for w in [5, 20]:
        base[f"close_ma_{w}"] = df["Close"].rolling(w).mean()
        base[f"close_std_{w}"] = df["Close"].rolling(w).std()

    df = pd.concat([df, base], axis=1).copy()
    return df

# =============================
# 1) 
# =============================
ticker = "TSM"
df = yf.download(ticker, start="2015-01-01", auto_adjust=False)
df.columns = df.columns.get_level_values(0)
df = df.reset_index()

df = add_features(df)

# =============================
# 2) next-day Open log-return level
# y = log(Open_{t+1}) - log(Open_t)
# =============================
df["target"] = np.log(df["Open"].shift(-1)) - np.log(df["Open"])
df = df.dropna().copy()

feature_cols = [c for c in df.columns if c not in ["Date", "target"]]
X = df[feature_cols]
y = df["target"]

# =============================
# 3) TimeSeriesSplit +  Grid8   3 folds = 24 
# =============================
tscv = TimeSeriesSplit(n_splits=4)

param_grid = {
    "n_estimators": [400, 800],      # 2
    "max_depth": [3, 4],             # 2  -> 2*2=4
    "learning_rate": [0.05],         # 1
    "subsample": [0.8],              # 1
    "colsample_bytree": [0.9],       # 1
    "min_child_weight": [1, 3],      # 2  -> 4*2=8
    "reg_lambda": [1.0, 3.0],        # 2  -> 8*2=16
    "reg_alpha": [0.0, 0.1, 0.5],    # 3  -> 16*3=48 combos
}


grid_list = list(ParameterGrid(param_grid))
print("Total param combos:", len(grid_list), " | folds:", 3, " | total fits:", len(grid_list) * 3)

best_params = None
best_score = float("inf")

for i, params in enumerate(grid_list, 1):
    fold_scores = []

    for tr_idx, va_idx in tscv.split(X):
        X_tr, X_va = X.iloc[tr_idx], X.iloc[va_idx]
        y_tr, y_va = y.iloc[tr_idx], y.iloc[va_idx]

        model = XGBRegressor(
            objective="reg:squarederror",
            random_state=42,
            tree_method="hist",
            n_jobs=-1,
            **params
        )
        model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False)
        pred = model.predict(X_va)
        fold_scores.append(rmse(y_va, pred))

    score = float(np.mean(fold_scores))
    print(f"[{i}/{len(grid_list)}] CV RMSE={score:.6f} params={params}")

    if score < best_score:
        best_score = score
        best_params = params

print("\nBest params:", best_params)
print("Best CV RMSE (log-return):", best_score)

# =============================
# 4) 
# =============================
final_model = XGBRegressor(
    objective="reg:squarederror",
    random_state=42,
    tree_method="hist",
    n_jobs=-1,
    **best_params
)
final_model.fit(X, y, verbose=False)

# =============================
# 5) 2024+  Actual vs Predicted
# =============================
df_future = yf.download(ticker, start="2015-01-01", auto_adjust=False)
df_future.columns = df_future.columns.get_level_values(0)
df_future = df_future.reset_index()

df_future = add_features(df_future)
df_future = df_future.dropna().copy()

#  
X_future = df_future.reindex(columns=feature_cols, fill_value=0.0)

pred_logret = final_model.predict(X_future)
open_today = df_future["Open"].values
pred_next_open = open_today * np.exp(pred_logret)

true_next_open = df_future["Open"].shift(-1).values
pred_next_open = pred_next_open[:-1]
true_next_open = true_next_open[:-1]

print("\nBacktest 2024+ (Open Price):")
print("MAE:", mean_absolute_error(true_next_open, pred_next_open))
print("RMSE:", rmse(true_next_open, pred_next_open))

plt.figure(figsize=(12, 5))
plt.plot(true_next_open, label="Actual", marker="o", linewidth=1)
plt.plot(pred_next_open, label="Predicted", marker="x", linewidth=1)
plt.title("Real Future Open Price vs Model Prediction (2024+ Backtest)")
plt.xlabel("Time Index")
plt.ylabel("Open Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# =============================
# 6)    Open
# =============================
last_row = df_future.iloc[-1]
X_latest = X_future.iloc[[-1]]

pred_lr = float(final_model.predict(X_latest)[0])
today_open = float(last_row["Open"])
pred_open_next = today_open * np.exp(pred_lr)

print("\nForecast (Next Trading Day):")
print("Today Open:", today_open)
print("Predicted next-day Open:", pred_open_next)
print("Predicted % change:", (pred_open_next / today_open - 1) * 100, "%")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

def rmse(a, b):
    return float(np.sqrt(mean_squared_error(a, b)))

def walk_forward_backtest(df_full, feature_cols, best_params,
                          first_test_date="2019-01-01",
                          retrain_months=6):
    """
    Walk-forward backtest:
    -  first_test_date 
    -  retrain_months  next-day open
    - 
    """
    df_full = df_full.sort_values("Date").copy()
    df_full["Date"] = pd.to_datetime(df_full["Date"])

    preds, trues, dates = [], [], []

    current_start = pd.to_datetime(first_test_date)
    last_date = df_full["Date"].max()

    while current_start < last_date:
        #  current_start 
        train_df = df_full[df_full["Date"] < current_start].copy()

        # current_start retrain_months
        current_end = current_start + pd.DateOffset(months=retrain_months)
        test_df = df_full[(df_full["Date"] >= current_start) & (df_full["Date"] < current_end)].copy()

        if len(train_df) < 300 or len(test_df) < 30:
            # /
            current_start = current_end
            continue

        X_train = train_df[feature_cols]
        y_train = train_df["target"]

        X_test = test_df[feature_cols]

        #  best_params
        model = XGBRegressor(
            objective="reg:squarederror",
            random_state=42,
            tree_method="hist",
            n_jobs=-1,
            **best_params
        )
        model.fit(X_train, y_train, verbose=False)

        #  log-return   next-day open
        pred_logret = model.predict(X_test)
        open_today = test_df["Open"].values
        pred_next_open = open_today * np.exp(pred_logret)

        #  next-day open next-day
        true_next_open = test_df["Open"].shift(-1).values

        preds.extend(pred_next_open[:-1])
        trues.extend(true_next_open[:-1])
        dates.extend(test_df["Date"].values[:-1])

        # 
        current_start = current_end

    return pd.DataFrame({"Date": dates, "Actual": trues, "Predicted": preds})


# ======= =======
FIRST_TEST_DATE = "2019-01-01"   # 2015-2018  2019 
RETRAIN_MONTHS = 6              #  6 

wf = walk_forward_backtest(
    df_full=df,
    feature_cols=feature_cols,
    best_params=best_params,
    first_test_date=FIRST_TEST_DATE,
    retrain_months=RETRAIN_MONTHS
)

print("Walk-Forward rows:", len(wf))
print("Walk-Forward MAE:", mean_absolute_error(wf["Actual"], wf["Predicted"]))
print("Walk-Forward RMSE:", rmse(wf["Actual"], wf["Predicted"]))
```

```python
import matplotlib.pyplot as plt

# =============================
# 1)  Static Backtest 
#  final_model
# =============================
df_static = df.copy()

X_static = df_static[feature_cols]
pred_logret_static = final_model.predict(X_static)

open_today_static = df_static["Open"].values
pred_next_open_static = open_today_static * np.exp(pred_logret_static)

true_next_open_static = df_static["Open"].shift(-1).values

#  next-day
dates_static = df_static["Date"].values[:-1]
pred_next_open_static = pred_next_open_static[:-1]
true_next_open_static = true_next_open_static[:-1]

# =============================
# 2) Walk-Forward wf
# =============================
dates_wf = wf["Date"]
pred_wf = wf["Predicted"]
true_wf = wf["Actual"]

# =============================
# 3) 
# =============================
plt.figure(figsize=(13,6))

# Static
plt.plot(
    dates_static,
    pred_next_open_static,
    color="orange",
    alpha=0.35,
    label="Static model prediction"
)

# Walk-forward
plt.plot(
    dates_wf,
    pred_wf,
    color="red",
    linewidth=1.5,
    label="Walk-forward prediction"
)

# Actual
plt.plot(
    dates_wf,
    true_wf,
    color="black",
    linewidth=1,
    label="Actual open price"
)

plt.title("Static vs Walk-Forward Backtest (Next-Day Open)")
plt.xlabel("Date")
plt.ylabel("Open Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
